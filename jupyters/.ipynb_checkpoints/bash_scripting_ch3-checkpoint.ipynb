## 1. IF statements

**Don't forget spaces between brackets and all the conditional elements inside (first line)**

### IF statements and string

````bash
# example
x="Queen"

if [ $x == "King" ]; then # don't forget spaces between brackets and all the conditional elements inside (first line)
    echo "$x is a king!"
else
    echo "$x is not a King!"
fi
````

`!=` not equal to

### Arithmetic IF statements (option 1)
Using double-parenthesis structure.

````bash
# Example
x=10

if (( $x > 5 )); then
    echo "$x is more than 5"
fi
````

### Arithmetic IF statements (option 2)
You can also use special flags for numerical comparisons using the square brackets.



````bash
# Same example but using [] and flags

x=10

if [ $x -gt 5 ]; then
    echo "$x is greater than 5"
fi
````

### Using AND and OR in Bash
You can also combine multiple conditions using AND notation as well as OR notation.



### Multiple conditions
To use multiple conditions, such as using an AND or an OR statement, you can either chain it and make multiple square-bracket conditions as follows.

#### Method 1: Chaining conditions with multiple square-bracket

````bash
# Example of Chaining conditions with multiple square-bracket
x=10

if [ $x -gt 5 ] && [ $x -lt 11 ]; then
    echo "$x is greater than 5 but less than 11"
fi
````

#### Method 2: Putting all conditions inside double-square-bracket notation `[[]]`

Double-square-bracket notation `[[]]` is used when we test for more than 1 condition.

````bash
x=10

if [[ $x -gt 5 && $x -lt 11 ]]; then
    echo "$x is greater than 5 but less than 11"
fi
````

### WHILE statement syntax

* WHILE loops are similar to FOR loops. 
* However, instead of a defined set to loop over, a WHILE loop sets a condition to test each iteration. 
* The iteration continues until this condition is no longer true.

````bash
x=1

# continue while x is less than or equal to three => Condition. 
# STOP the iteration when the condition is no longer met / condition no longer true
while [ $x -le 3 ]; #=> condition
do
    echo $x
    ((x+=1)) # increase x by one using the double parenthesis
done
````

* The variable x is set to one to start. 
* The condition in square brackets `[ condition ]` is to continue while x is less than or equal to three. 
* At each iteration, the value of x is printed out. 
* Then x is increased by one using the double parenthesis arithmetic method `(arithmetic)` you previously learned about. 

As expected, the numbers one to three are printed out.

## 3. CASE statements

[link](https://campus.datacamp.com/courses/introduction-to-bash-scripting/control-statements-in-bash-scripting?ex=8)

CASE statements help create clearer and more efficient code in cases of complex or multiple (perhaps nested) IF statements. Consider the following example of conditionals and actions for a data processing pipeline.



### A complex IF statement

You feed in the filename as the first ARGV argument. 

Then grep checks **inside** for the words noted previously. Doesn't this seem complex and repetitious?



````bash
#!/usr/bin/env bash

# File given by the user. Storing it in a variable
file=$1

# I'm want to search for particular words and do some actions if I find them
# inside the file

# looking in the contents of the file for the word "sydney"
if grep -q "sydney" $file; then
    mv $file sydney/
fi

# looking in the contents of the file for the word "melbourne" o "brisbane"
if grep -q "melbourne|brisbane" $file; then
    rm $file
fi

# looking in the contents of the file for the word "canberra"
if grep -q "canberra" $file; then
    mv $file "IMPORTANT_$file"
fi
````

Doesn't this seem complex and repetitious?

Let's break down the basic CASE statement structure format on the right. 

The STRINGVAR will be checked against each PATTERN and if it matches then the associated code will be run. 





The last two parts of the CASE statement are as follows. 

**The default match.** It is common, though not required, to have a default condition by matching on just a star symbol `*`. 

If all other pattern matching fails, the code for the default match will run.



Finally, finish with the word 'esac'. Which is CASE backwards! 

### From IF to CASE

Let's see the old IF statement previously created here. We can now turn it into a CASE statement. 

* We assume the first ARGV element is a filename which we concatenate into the variable to match against.
* We can use stars `*` on either side of the words as we are regex matching, so the text can have anything before or after our desired word.
* At the end, we add a default condition (if any of the patterns was found inside the contents of the file) printing out that it wasn't recognized, no cities / patterns were found. 


````bash
#!/usr/bin/env bash

# File given by the user. Storing it in a variable
file=$1

# case $file in <- wrong!

# correct way:
case $(cat $file) in
    *sydney*)
    mv $file sydney/ ;;
    *melbourne*|*brisbane*)
    rm $file ;;
    *canberra*)
    mv $file "IMPORTANT_$file" ;;
    *)
    echo "No cities found" ;;
esac
````

### Days of the week with CASE - exercise

In your role as a Data Scientist, it is sometimes useful to associate dates with a 'working day' (Monday, Tuesday, Wednesday, Thursday, Friday) or a 'weekend' (Saturday & Sunday).

Your task is to build a small Bash script that will be useful to call in many areas of your data pipeline. It must take in a single argument (string of a day) into ARGV and use a CASE statement to print out whether the argument was a weekday or a weekend. You only need to account for the capitalized case for now.

You also don't need to worry about words or letters before and after. Just use exact matching for this example.

Remember the basic structure of a case statement is:

```
case MATCHVAR in
  PATTERN1)
  COMMAND1;;
  PATTERN2)
  COMMAND2;;
  *)
  DEFAULT COMMAND;;
esac
```

**Instructions**

* Build a CASE statement that matches on the first ARGV element.
* Create a match on each weekday such as `Monday`, `Tuesday` etc. using OR syntax on a single line, then a match on each weekend day (`Saturday` and `Sunday`) etc. using OR syntax on a single line.
* Create a default match that prints out `Not a day!` if none of the above patterns are matched.
* Save your script and run in the terminal window with `Wednesday` and `Saturday` to test.

````bash
# Datacamp answer

# Create a CASE statement matching the first ARGV element
case $1 in
  # Match on all weekdays
  Monday|Tuesday|Wednesday|Thursday|Friday)
  echo "It is a Weekday!";;
  # Match on all weekend days
  Saturday|Sunday)
  echo "It is a Weekend!";;
  # Create a default
  *) 
  echo "Not a day!";;
esac
````

This case statement simplifies what could have been a lengthy few IF statements.

You are working as a data scientist in charge of analyzing some machine learning model results. The production environment moves files into a folder called `model_out/` and names them `model_RXX.csv` where XX is a random number related to which experiment was run.

**Instructions**

* Use a FOR statement to loop through (using glob expansion) files in `model_out/`.
* Use a CASE statement to match on the contents of the file (we will use cat and shell-within-a-shell to get the contents to match against). It must check if the text contains a tree-based model name and move to `tree_models/`, otherwise delete the file.
* Create a default match that prints out `Unknown model in FILE` where FILE is the filename then run your script.


````bash
ls ../data/model_out_ex/
````

````bash
for file in ../data/model_out_ex/model_out/* ;
do
# cat to get the contents of each file and try to match the patterns i'm looking for
    case $(cat $file) in
        "Random Forest"|GBM|"XGBoost")
        mv $file ../data/model_out_ex/tree_models/ ;;
        KNN|Logistic)
        rm $file ;;
        *)
        echo "Unknown model in $file" ;;
    esac
done
````

**What was wrong?**

I only wrote `"Random Forest"|GBM|"XGBoost"` and `KNN|Logistic`. Apparently, these words were never alone but had characters before and after them. So writing the words alone will not match anything.

I had to use * to match these words even if they had characters before and after them:

`*"Random Forest"*|*GBM*|*"XGBoost"*` and `*KNN*|*Logistic*`

````bash
for file in ../data/model_out_ex/model_out/* ;
do
# cat to get the contents of each file and try to match the patterns i'm looking for
    case $(cat $file) in
        *"Random Forest"*|*GBM*|*"XGBoost"*)
        mv $file ../data/model_out_ex/tree_models/ ;;
        *KNN*|*Logistic*)
        rm $file ;;
        *)
        echo "Unknown model in $file" ;;
    esac
done
````

````bash
count=0

for file in ../data/model_out_ex/model_out/* ;
do
# cat to get the contents of each file and try to match the patterns i'm looking for
    case $(cat $file) in
        *"Random Forest"*|*GBM*|*"XGBoost"*)
        mv $file ../data/model_out_ex/tree_models/ 
        ((count=count+1)) ;;
        *KNN*|*Logistic*)
        rm $file ;;
        *)
        echo "Unknown model in $file" ;;
    esac
done

echo "There were $count tree-based models"
````
