## Basic variables in Bash

```bash
var1="Moon"
echo $var1
```

```bash
firstname="Gloriem"
lastname="Perezim"

echo "Hi there" $firstname $lastname
```

### Single, double, backticks

 The last way is quite special and powerful. 
 
 Backticks creates a shell within a shell. 
 
 That is, it sends whatever is between the backticks out to a new shell and **takes the results back into the variable**. 
 
 Let's see these methods in action. 

Let's see the effect of different types of quotation marks for variable creation. 

### Different variable creation

```bash
now_var='NOW'
now_var_singlequotes='$now_var'

echo $now_var_singlequotes
```

You can see that the use of single quotes treats what we gave it literally, it doesn't understand there is text inside that variable (`$now_var`).

 We also create a variable with double quotes in the same way. 

```bash
now_var_doublequotes="$now_var"

echo $now_var_doublequotes
```

The double quotes understands our variable call and returns the NOW text inside that variable (`$now_var`). 

### The date program
We will use the date program to demonstrate the use of backticks for variable creation.

```bash
date
```

```bash
echo `date`
```

#### Shell within a shell


Now let's use the powerful shell-within-a-shell construct. 

We create a variable similar to before. However, instead of using single quotes, we use use backticks around the word date and echo the resulting variable. 

```bash
rightnow_doublequotes="The date is `date`"

echo $rightnow_doublequotes
```

We can see the result includes the current date and time. 

What actually happened here is that a mini shell was opened between those backticks and called the date program. 

The result of this call was then saved back as a string in-line. 

### Parentheses vs backticks

There is an alternate way to invoke a shell-within-a-shell. 

This uses a dollar sign around parenthesis as can be seen here `$(command)`. 

We create two variables around the date command and echo them both. 


```bash
# changing the variable names :D 
dolar_parentheses="The date is $(date)"
backticks="The date is `date`"

echo $dolar_parentheses
echo $backticks
```

You can see the result is exactly the same. Either method works. 

The backticks method is more backwards compatible as it is older. 

Though the parenthesis method is more modern and has some advantages, discussed in the link here.

```bash
expr 5 + 7
```

### Shell within a shell revisited

```bash
model1=87.65
model2=89.20

echo "$model1 + $model2"
echo "$model1 + $model2" | bc
echo "The total score is $(echo "$model1 + $model2" | bc)"

echo ""
echo "Doing something a little more complex:"

echo "($model1 + $model2)/2"
echo "($model1 + $model2)/2" | bc
echo "The average score is $(echo "($model1 + $model2)/2" | bc)"
```

### Converting Fahrenheit to Celsius - exercise
You work in the analytics department for an Australian company that recently purchased an American company. The files and data from the US company are in the imperial system and need to be converted to metric. This sounds like a great opportunity to use your Bash skills to create a program which will assist.

Your task is to write a program that takes in a single number (a temperature in Fahrenheit) as an ARGV argument, converts it to Celsius and returns the new value. There may be decimal places so you will need to undertake calculations using the `bc` program.

* At all times use 2 decimal places using the `scale` command for `bc`.

The formula for Fahrenheit to Celsius is:

`C = (F - 32) x (5/9)`

Remember, since we are using arguments, you will need to run your script from the terminal pane rather than 'run this file' button.

**Instructions**:

* Create a variable `temp_f` from the first ARGV argument.
* Call a shell-within-a-shell to subtract 32 from `temp_f` and assign to variable `temp_f2`.
* Using the same method, multiply `temp_f2` by 5 and divide by 9, assigning to a new variable `temp_c` then print out `temp_c`.
* Save and run your script (in the terminal) using 108 Fahrenheit (the forecast temperature in Parramatta, Sydney this Saturday!).


```bash
# testing
temp_f=108

# Get first ARGV into the variable

# Subtract 32
echo "scale=2; $temp_f - 32" | bc
# assign to variable temp_f2
temp_f2=$(echo "scale=2; $temp_f - 32" | bc)
echo "The value of the variable temp_f2 is $temp_f2"

echo ""
echo "scale=2; $temp_f2*5/9" | bc
# assigning to a new variable temp_c
temp_c=$(echo "scale=2; $temp_f2*5/9" | bc)
echo "The value of the variable temp_c is $temp_c"

echo ""
echo "Doing all the operations at once"
echo "scale=2; ($temp_f-32)*5/9 => passing this to bc command" 
echo "scale=2; ($temp_f-32)*5/9" | bc
temp_c2=$(echo "scale=2; ($temp_f-32)*5/9" | bc)
echo "$temp_c2"
echo "The value of the variable temp_c2 is $temp_c2"
echo "$temp_f°F is, in Celsius, $temp_c2"
```

```bash
#!/usr/bin/env bash

# Get first ARGV into the variable
temp_f=$1

# Subtract 32
temp_f2=$(echo "scale=2; $temp_f - 32" | bc)

# Multiply by 5/9 and print
temp_c=$(echo "scale=2; $temp_f2*5/9" | bc)
#echo "The value of the variable temp_c is $temp_c"

# Print the celsius temp
echo "$temp_c"
```

## Arrays in Bash

Unlike many programming languages, Bash separates array elements with spaces and not commas.

```bash
# Incorrect:
my_array=(1,2,3)
# Correct
my_array=(1 2 3)
```

### Useful array properties

```bash
my_array=(1 10 34 25 49 1000)

# You can return all array elements by indexing into the array using the @ symbol.
echo "The elements of my bash array are:" ${my_array[@]}

# The length of an array is returned by adding a pound symbol (#) to the front of a call to the entire array. 
echo "The length of my bash array is" ${#my_array[@]}
# Exactly the same as the previous but with # before the variable name
```

### Manipulating array elements


```bash
# to return the 3rd element of my array
echo ${my_array[2]}
```

#### Change array values using index notation


```bash
my_array2=(232 214214 32434 2113 56546 213213 56546 1212 121 656 121 43543 121 156)
# the 8th element
echo "The original value of the 8th element of my array2 is: "${my_array2[7]}
# changing the value of the 8th element
my_array2[7]=999
echo "The new value of the 8th element of my array2 is: " ${my_array2[7]}
```

#### Slice out parts of an array



The notation adds to the call to **print the entire array** with a colon, number, colon, number.

`N`: means the starting index and 

`M`: means the number of elements to return after that index. 

In the example above, we start at the third index (or fourth element) and return 2 elements. 

```bash
# working on my_array2
echo "My array has a total of" ${#my_array2[@]} "elements"
echo "The elements of my_array2 are" ${my_array2[@]}
# slicing my array
echo "A slice of my array starting at the 3 index (4th element) and returning 4 elements in total: " ${my_array2[@]:3:4}
```

#### Appending to arrays


`array+=(elements to append)`

The elements you want to add are surrounded by parentheses. 

```bash
my_array3=(232 214214 32434 2113 56546 213213 56546 999 121 656 121 43543 121 156)
echo "The original elements of my array: " ${my_array3[@]}
echo "Appending 1024 at the end of my array ..."
my_array3+=(1024)
echo "The new element, 1024, should be appended at the end of my array: " ${my_array3[@]}
```

### Associative arrays


#### Creating an associative array

Important:

* Associative arrays allow you to index using words rather than numbers, which can be important for ease of inputting and accessing properties.

```bash
# Declare first
declare -A city_details
# Add elements
city_details=([city_name]="New York" [population]="12000000")
```

 We declare first using the correct flag and call the array city_details. 
 
 We set the keys to be the 'city_name' and the 'population' with these associated values. 
 
 We can index into this array using any of the keys. 
 
 Here we index using the `city_name` key and it returns what we expect. New York. 

```bash
# Index using a key to return a value
echo ${city_details[city_name]}
```

```bash
# Another method for creating an associative array: declare and add elements all on the same line
declare -A city_details2=([city_name]="Neo Tokyo" [population]=1000)

# testing -> indexing using the keys
echo "The name of the city is" ${city_details2[city_name]} "and its population is" ${city_details2[population]}
echo "The name of the city is ${city_details2[city_name]} and its population is ${city_details2[population]}"
```

#### Access the 'keys' of an associative array using `!` in `${!my_array[@}`

```bash
# Access the 'keys' of an associative array using !
echo "The keys of this associative array are:" ${!city_details2[@]}
echo "The keys of this associative array are: ${!city_details2[@]}" 
```
